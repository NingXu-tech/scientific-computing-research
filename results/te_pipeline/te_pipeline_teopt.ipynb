# TE pipeline (te_opt version)
This notebook imports `te_opt.te_runner` and runs TE + export.


```python
import pandas as pd
import numpy as np
from pathlib import Path
from te_opt.te_runner import TERunConfig, run_multivariate_te, export_te_results

OUT_DIR = Path('te_opt/out_from_notebook')
OUT_DIR.mkdir(parents=True, exist_ok=True)

cfg = TERunConfig(verbosity=0)
```

```python
# Example: analytic-ex1
df = pd.read_csv('./dataset/analytic-ex1.dat', delim_whitespace=True)
data_array = df.iloc[:200].values.T

res = run_multivariate_te(data_array, cfg=cfg)
export_te_results(res, out_dir=OUT_DIR, prefix='analytic-ex1_demo', fdr=False)
print('DONE:', OUT_DIR)
```
